```python
# libraries
import rasterio  
import bokeh
import urllib.request
import zipfile
import os
import numpy as np
import pandas as pd
import geopandas as gpd

import matplotlib as mlp
from rasterio.plot import show
from rasterio.windows import Window
import matplotlib.pyplot as plt
%matplotlib inline

from esda.moran import Moran
from libpysal.weights.contiguity import Queen
import libpysal
import warnings
warnings.filterwarnings('ignore')
```

```python
def getPopulationOfYear(turkey, years , category):  # turkey location, list of years, category (total, rural, urban), path to dataset
    
    totalPopulation = []   # array for population of each year
    
    for year in years:  # loop over all years in years array
        
        if(category == 'Total'):
            ssp4_tif = rasterio.open(os.path.join("ssp4_"+year+".tif"))  # path for Total Dataset
        elif(category == 'Urban'):
            ssp4_tif = rasterio.open(os.path.join("ssp4urb"+year+".tif"))  # Path for Urban Dataset
            
        pop = ssp4_tif.read(1)   # read population data for each year
        pop = pop.astype('int64')  # convert datatype or population to int64 to avoid overflow
        pop[pop < 0] = 0   # set negitive values to zero
        totalPopulation.append( np.sum(pop[turkey])) # appending population of year year to list above
        
    return totalPopulation # return list of population for all years
```

```python
years = [str(i) for i in range(2010, 2110, 10)]  # list of years from 2010 to 2100 with jump of 10 years
years
```

```python
# get location for Turkey
countries_tif = rasterio.open('countries.tif')
countries = countries_tif.read(1)

#for Turkey
turkey = countries == 792 # = Turkey
```

```python
# Total population of turkey from 2010 to 2100 from 
totalPopulation = getPopulationOfYear(turkey, years, "Total")
```

```python
# Urban population of turkey from 2010 to 2100 from 
urbanPopulation = getPopulationOfYear(turkey, years, "Urban")
```

```python
# Data
x = [i for i in range(2010, 2110, 10)]  # generating years

 
# multiple line plot
plt.rcParams["figure.figsize"] = (18,8)
plt.plot(x, totalPopulation, label = "Total Population")
# plotting the line 2 points 
plt.plot(x, urbanPopulation, label = "Urban Population" )
plt.xlabel('Years', fontsize=15)
# Set the y axis label of the current axis.
plt.ylabel('Population in Millions', fontsize=15)
# Set a title of the current axes.
plt.title('Total and Urban Population of Turkey from 2010 to 2100' , fontsize=15)
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

```python
pop_2010_tiff = rasterio.open(os.path.join("ssp4_2010"".tif"))  
pop_2100_tiff = rasterio.open(os.path.join("ssp4_2100"".tif"))
```

```python
pop_2010 = pop_2010_tiff.read(1)   # read population data for each year
pop_2010 = pop_2010.astype('int64')  # convert datatype or population to int64 to avoid overflow
pop_2010[pop_2010 < 0] = 0   # set negitive values to zero
```

```python
pop_2100 = pop_2100_tiff.read(1)   # read population data for each year
pop_2100 = pop_2100.astype('int64')  # convert datatype or population to int64 to avoid overflow
pop_2100[pop_2100 < 0] = 0   # set negitive values to zero
```

```python
delta_pop = pop_2100 - pop_2010   # difference between population in 2100 to 2010
```

```python
delta_pop[delta_pop > 0] = 0 #set values of increase in population to zero
```

```python
delta_pop = delta_pop * -1  # assign positive sign for calculation
```

```python
delta_pop = delta_pop.astype(dtype=rasterio.ubyte)    # changing data type to raterio ubyte in order to draw raster image
```

```python
with rasterio.open(
        'reduced_population_cells.tif', 'w',
        driver='GTiff', width=500, height=300, count=1,                   #saving image to disk 
        dtype=delta_pop.dtype) as dst:
    dst.write(delta_pop, window=Window(50, 30, 250, 150), indexes=1)
```

```python
reduced_population_cells_tif = rasterio.open('reduced_population_cells.tif')            #reading the saved image to plot
```

```python
plt.figure(figsize=(14, 14))
show((reduced_population_cells_tif, 1), title='Areas projecting reduction in Population from 2010 to 2100')
```

```python
admin_country_df = gpd.read_file("ne_110m_admin_0_countries.shp") # reading SHP file
admin_country_df.head(5) # first 5 rows
```

```python
mortality_df = pd.read_csv('under5mortality.csv') # reading csv file for mortality
mortality_df.head(5) # first 5 rows
```

```python
result_df = pd.merge(admin_country_df,mortality_df, left_on = "ISO_A3", right_on= "ISO", how= "left") # performing left join
result_df
```

```python
spatial_weights = Queen.from_dataframe(result_df)    # calculating spatial weights
```

```python
result_df['ChildMortality'] = result_df['ChildMortality'].fillna(0)  # set Nan values to zero
```

```python
moran = Moran(result_df['ChildMortality'], spatial_weights)    # calculating Moran I value
```

```python
print(f'Moran I Value is : {moran.I}')
```
